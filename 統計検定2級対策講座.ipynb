2. 付表を使えるように整理: $z=(\bar{x}-\mu)/(\sigma/\sqrt{n})\sim N(0,1)$
3. 付表で$P(-■\le z\le ■)=0.95$となる$■$を探して
$\bar{x}-■\sigma/\sqrt{n} \le \mu\le\bar{x}+■\sigma/\sqrt{n}$

### 母平均の区間推定(1標本問題)


正規母集団$N(\mu,\sigma^2)$の母平均$\mu$の信頼区間

母分散($\sigma^2$)が既知

正規分布で$P(-■\le z\le■)＝0.95$
$\bar{x}-■ \frac{\sigma}{\sqrt{n}}\le \mu \le \bar{x}+■\frac{\sigma}{\sqrt{n}}$


- STEP1: 分布
$x\sim N(\mu, \sigma^2)$ $\bar{x}\sim(\mu, \sigma^2/n)$
- STEP2: 整理
$z=(\bar{x}-\mu)/(\sigma/\sqrt{n})\sim N(0,1)$
- STEP3: 確率計算
$P(-■\le z\le■)　＝0.95$
$\bar{x}-■ \sigma/\sqrt{n}\le \mu \le \bar{x}+■\sigma/\sqrt{n}$



母分散($\sigma^2$)が未知

自由度$n-1$の$t$分布で$P(-■\le t\le■)　＝0.95$
$\bar{x}-■ \hat{\sigma}/\sqrt{n}\le \mu \le \bar{x}+■\hat{\sigma}/\sqrt{n}$


- STEP1: 分布
$\bar{x}\sim(\mu, \sigma^2/n)$
- STEP2: 整理
$t=(\bar{x}-\mu)/(\hat{\sigma}/\sqrt{n})\sim t(n-1)$
- STEP3: 確率計算
$P(-■\le t\le■)　＝0.95$
$\bar{x}-■ \hat{\sigma}/\sqrt{n}\le \mu \le \bar{x}+■\hat{\sigma}/\sqrt{n}$

### 母分散の区間推定(1標本問題)


正規母集団$N(\mu,\sigma^2)$の母分散$\sigma^2$の信頼区間

自由度$n-1$の$\chi^2$分布で$P(-△\le w\le▲)＝0.95$
$\frac{(n-1)\hat{\sigma}^2}{▲} \le\sigma^2 \le \frac{(n-1)\hat{\sigma}^2}{△}$


- STEP1:
$x_1,\cdots,x_n\sim N(\mu,\sigma^2)$
$x_1,\cdots,x_n\sim N(0,1) \larr z_i = \frac{(x_i-\mu)}{\sigma}$
- STEP2: 自由度$n-1$の$\chi^2$分布
$w =\sum{z_i^2} = \frac{(n-1)\hat{\sigma}^2}{\sigma^2}\sim \chi^2(n-1)$
- STEP3:
$P(-△\le w\le▲)＝0.95$
$\frac{(n-1)\hat{\sigma}^2}{▲} \le\sigma^2 \le \frac{(n-1)\hat{\sigma}^2}{△}$


### 母比率の区間推定(1標本問題)


母比率(サンプルサイズが大きい前提)の信頼区間

正規分布で$P(-■\le z\le■)＝0.95$
$\hat{p}-■\sqrt{\frac{\hat{p}(1-p)}{n}}\le p \le\hat{p}+■\sqrt{\frac{\hat{p}(1-p)}{n}} ※pは\hat{p}で近似$

- STEP1:
$n人中x人がYes \rarr x\sim B(n,p)$
$中心極限定理よりx\sim N(np,np(1-p))$
$標本比率 \hat{p}=\frac{x}{n}\sim \frac{N(p,p(1-p)}{n}$
- STEP2:
$z =\frac{\hat{p}-p}{\sqrt{p(1-p)/n}} \sim n(0,1)$
- STEP3:
$P(-■\le z\le■)＝0.95$
$\hat{p}-■\sqrt{\frac{\hat{p}(1-p)}{n}}\le p \le\hat{p}+■\sqrt{\frac{\hat{p}(1-p)}{n}} ※pは\hat{p}で近似$
